# Question #1 -- Grayson's Response
1. This paper highlights the importance of data tidying and the advantages of a consistent data structure like Tidy datasets. Although we devote so much time to cleaning data, the topic is largely unexplored in research. Tidy datasets' specific structure makes them easier to manipulate and analyze, allowing us to develop "tidy tools" that can effectively  analyze messy datasets.

2. The tidy data standard standardizes data exploration and data cleaning. It aims at organizing values within a dataset to simplify and faciliate analysis and the development of tidy tools.

3. This first sentence compares messy/clean datasets to family dynamics. So, tidy datasets are all structured and streamline cleaning and analysis processes. Meanwhile, messy datsets have complex, unique  problems inherent to distinct datasets. With the regard to the second sentence, when you look at a specific dataset, it is typically easy to identify the observations and variables. However, in general, it can be difficult to define both observations and variables because datasets are often structured in many different ways, depending on who created the dataset.


4. Values are the individual pieces of data themselves, classified typically as numbers or strings. They are organized by variables and observation. A variable is all of the value under the same "attribute" or category, such as temperature or height. Meanwhile, Wickham defines an observation as contains all of the values of the same unit, such as a person or treatment, across attributes.

5. Tidy data is defined by a specific structure: each variable is a column, each observation is a row, and each type of observational unit is a table.

6. The 5 most common problems with messy datasets are that 1) column headers are values, not variable names 2) multiple variables are stored in one column 3) variables are stored in both rows and columns 4) multiple types of observational units are stored in the same table and 5) a single observational unit is stored in multiple tables. The data in Table 4 are messy because both the rows and columns are variables and the column headers are values instead of variable names. To tidy a dataset like this, we can "melt" it, which is the process of turning columns (that are already variables) into rows.

7. Table 11 is messy because it uses variables as columns (d1, d2, d3...). Each day is technically an observation/value. There is also a lot of missing data. On the other hand, Table 12a is  molten because the days, which were columns in Table 11, are now values under the "date" column. However, it's not fully tidy yet because the column "element" is missing actual values. Table 12b is fully tidy because tmax and tmin are now column names with specific values, not acting as a variable.

8. The chicken-and-egg problem is that "if tidy data is only as useful as the tools that work with it, then tidy tools will be inextricably linked to tidy data." Wickham hopes to use human factors, user-centered design, ethnography, etc. to improve, holistically, our understanding "of the cognitive side of data analysis" and hopes to improve how we work with and design tidy tools.
