```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Q2 Pt. 1
airbnb_url = 'https://raw.githubusercontent.com/DS3001/wrangling/main/assignment/data/airbnb_hw.csv'
df = pd.read_csv(airbnb_url)
```

```python
print(df.shape)
print(df.columns)
df.head()
```

```python
print(df.dtypes)
```

```python
df['Price'].unique()

# 'Price' column is not classified as an integer because there are non-integer
# values (commas) once the price exceeds 999 dollars --> why they are strings
```

```python
price = df['Price']
price1 = price.str.replace(',','')
print(price1.unique())
```

```python
price = pd.to_numeric(price1,errors='coerce')
print(price.unique())

# Typecast from string to integer
```

```python
df['Price_Missing'] = df['Price'].isnull()
print('Missing data:', df['Price_Missing'])
```

```python
#Q2 Pt. 2
sharks_url = 'https://raw.githubusercontent.com/DS3001/wrangling/main/data/sharks.csv'
df2 = pd.read_csv(sharks_url, low_memory= False)
```

```python
print(df2.shape)
print(df2.columns)
df2.head()
```

```python
print(df2['Type'].unique())
print(df2['Type'].value_counts())
```

```python
type = df2['Type']
type = type.replace(['Sea Disaster', 'Boat','Boating','Boatomg'],'Watercraft')
type.value_counts()
type = type.replace(['Invalid','Questionable','Unconfirmed','Unverified','Under investigation'],'In Question')
type.value_counts()
df2['Type'] = type
df2['Type'].value_counts()
```

```python
df2['Type'] = df2['Type'].isnull()
print('Missing data:', df2['Type'])
```
